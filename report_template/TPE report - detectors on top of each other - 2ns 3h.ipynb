# Tandem Piercer Experiment - Measurement report

Load Python libraries:

```python
import os
from tpe.Stats import Stats
from tpe.functions import md, load_styles, get_report_header
```

Load inline graphics utility and output current working directory name:

```python
%pylab inline
print("\nCurrent directory: %s" % os.getcwd())
```

Initialize statistic class and output experiment statistic files to use in this report:

```python
st = Stats()
st.print_experiment_stat_files("../experiments/default_2022_2_1[5,6]*");
```

Define a working experiment directory and get a report header:

```python
directory = "../experiments\default_2022_2_15_21_31"
get_report_header(directory)
```

### Experiment statistics

Read statistics from the experiment and output basic data:

```python
st.read_stats_dataframe(directory);
st.print_basic_data();
```

## Detector clicks in time

```python
st.plot_channel_counts(sec=60)
```

## Detector accumulated clicks and sample rates

```python
st.plot_channel_totals_and_rates()
```

## Detectors coincidence counts and rates

```python
st.plot_coincidences_and_sample_rate()
```

```python
st.plot_coincidences_and_sample_rate(elapsed = (4535, 4540))
```

## Coincidence time histrogram and scatter plot

Set calibrated values, ADC corresponding to kilo-electronvolts:

```python
st.adc_calibrate_a(12500, 122)
st.adc_calibrate_b(12500, 122)
```

```python
st.plot_time_histogram_and_scatter()
```

### Only coincidences where time difference is zero

```python
st.plot_time_histogram_and_scatter(time_difference=100)
```

## Coincidence spectra

```python
st.plot_spectra(coincidences=True);
```

## Channel spectra

```python
st.fit_spectra(coincidences=False, log=True)
```

#### Data

```python
st.print_stats_link()
```

Output head and tail of the statistics data:

```python
st.stats
```
