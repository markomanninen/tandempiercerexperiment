American electrical engineer Eric Reiter originally invented a fundamental physics experiment with gamma-rays in 2001. I call it the Tandem Piercer Experiment in the following report.

Tandem Piercer Experiment (TPE) is based on Reiter's long research and study of the semi-classical electron model, the so-called threshold model. A similar model was called the accumulation hypothesis or the loading theory in early 20th-century physics history. TPE demonstrates how two scintillation detectors placed in tandem geometry together with gamma-ray sources can produce an effect that may either support the threshold model or indicate an unexplained anomaly in fundamental particle physics.

Max Planck and contemporary physicists first proposed the loading theory for electrons. Those particular ideas are now known as his second theory. Later on, the same scientists abandoned the loading theory, which eventually caused the loading theory to sink into oblivion in mainstream physics. The sudden discontinuous quantum jump of the electron in its energy states around the atom with no need of evolution in time was thought to be the best explanation instead of the pre-loaded state of the electron.

The TPE experiment and its result raises previous ideas in the daylight again and forces us to ask if there is still room for semi-classical interpretation in fundamental particle physics. TPE resembles the beam split coincidence experiment in certain ways. TPE also has implications in the wave-particle duality theory. Thus, I propose that this experiment resides in the middle of classical physics and early questions in quantum mechanics. The TPE is a fruitful real-world experiment to study and learn both areas of physics, ontic and epistemic questions related to them, as well as appreciate experimental science in general.

The second step is to measure the chance rate of the gamma-ray photopeaks. The third step is to measure the background rate. The final result is called Unquantum Effect Ratio or UQE in short. Unquantum-term is used because the chance rate measured in the second step shows the statistical prediction of the coincident gamma-ray photopeaks according to standard quantum mechanics. The result of the Tandem Piercer Experiment, on the other hand, yields a correlation between the detectors, not just random distribution of coincidences equal to chance. If UQE is greater than one, the coincidence rate of gamma-ray photopeaks exceeds the chance rate. The central hypothesis of the experiment is that it should not happen in the quantum world because the electron ionization in the scintillator produced by a single gamma-ray is expected to consume most of its energy in the process. There should be no energy enough for the second full-height gamma-ray photopeak.

For some reason, we often witness contrary in the Tandem Piercer Experiment. There seem to be more gamma-ray photopeak coincidences between the detectors than expected. It suggests that there is energy left from the photoelectric effect of the gamma-ray in the front detector for the second photoelectric effect in the rear detector, maybe even more. How could that be explained with a quantized particle/quantum model? Could the wave model of electromagnetic radiation with a resonating pre-loaded state of the electron explain it better? ## Step 1 - Detect gamma-ray photopeak

The first step is a crucial part of the experiment. I need to produce the pulse height spectrum (PHS) of the selected isotope and pinpoint the gamma-ray photopeak (later called just gammas) from it. I rather think of gammas as clicks in the detector without committing too much, at this point, what is the physical ontology of the gamma-ray in the electromagnetic spectrum - might it be a wave, a particle, a wave-particle, or a quantum oscillation in the electromagnetic field.

By scientists' previous work in history, we know that gammas reside in a certain energy level (keV) for specific isotopes. Reiter has found two isotopes that can be used in the Tandem Piercer Experiment, namely `Cadmium 109` and `Cobalt 57`. They have the clearest and most distinguished photopeaks in the Thallium doped Sodium Iodide, shortly `NaI(Tl)` scintillators. Corresponding gammas for `Cadmium 109` and `Cobalt 57` can be found from `88keV` and `122keV` respectively (https://www.cpp.edu/~pbsiegel/bio431/genergies.html).

The second criterion among the clear photopeak are the half-life time, which must also be reasonable to make the experiment possible. The half-life for `Cadmium 109` and `Cobalt 57` are `453 days` and `279 days`, suitable for doing the experiment over a few years.

The third criterion is that the spontaneous decay of an isotope must emit only one gamma-ray per decay. The fourth criterion is that the gamma must be in a part of the spectrum for the chosen detector to respond primarily by the photoelectric effect instead of the Compton effect. For example, `Cesium 137` on a `NaI(Tl)` scintillator responds primarily from the Compton effect and would make `Cs-137` a poor choice.

In this experiment, the following source is used:

<p class="source-title title">{pulse_source}</p>

The first part is the name of the isotope. The second part is the radiation activity value in microcuries (μCi). 1μCi means 3.70 × 10<sup>4</sup> disintegrations per second, which is another way of denoting the radionuclide activity in Becquerels (Bq). Exemption values declared by STUK limit that, as a consumer, I can use radionuclides up to 10μCi (10<sup>6</sup>Bq to be precise) to handle it safely. Radiation with a 10μCi source is practically zero when measuring radiation ten centimeters from the source. Thus, the isotope source needs to be set very close to the detectors, within a few centimeters.

Pulse signals from the detector are voltage variations converted from the analog to the digital values (ADC) in the oscilloscope. The pulse height spectrum comprises the maximum value of the triggered pulse signals retrieved from the oscilloscope. Maximum value counts are collected and plotted to the graph, usually in logarithmic scale, which shows how values start to accumulate as peaks in the graph.

In my graphs, I use Volts in the x-axis, which are ADC values provided by PicoScope converted to voltage values. Digital values correspond to the ADC-channel number in the above picture. So, in practice, there are two common ways of denoting the x-axis: ADC-channel numbers and Volts.

*Note: You should not confuse the above pulse height spectrum's channel-term with the A, B, C, D channels I am using to refer to the four different signal channels in the oscilloscope.*

KeV (kilo-electronvolt) could be used as a converted unit if the gamma-ray photopeak is manually defined to its location in the x-axis. The process is called _calibration_, which can also include comparing two or more different isotope sources to ensure that the system specifies a correct region of interest (ROI). Gamma-ray photopeak keV-energy (the energy of the incident radiation) is estimated to be linearly proportional to the emitted scintillation light energy and the electrical pulse energy produced by the photomultiplier.

There are also Multichannel Analyzer (MCA) devices in the market that could do the PHS. I have done the same with Python by getting the maximum value of each signal data buffer when PicoScope has triggered and found a pulse coming from the raw signal of the detector. In practice, getting the PHS in my setup is a combination of triggering the SCA square wave pulse and, at the same time retrieving data from the corresponding raw pulse signal.

__About the SCA module__

The SCA module detects the pulse from the detector and creates a high precision square pulse. The SCA module's square pulse height and width are always the same, while the curve of the pulses in the raw signal has various shapes, lengths, and heights. I rely on the SCA nuclear instrumentation (NIM) module in my experiment setup because it is dedicated to analyzing gamma-ray and x-ray pulses. SCA provides a square pulse with high precision up to picoscale, which means that I have a good starting point on measuring the coincident pulse distance between the two different channels, namely channels A and B.

### Regarding the resolution and time window

I will need information about the coincident timestamps to generate a time difference histogram in the following measurements. I use the histogram to confirm that coincidences have centered peaks in the graph. The time difference is measured by getting the triggered starting point of the square wave when they appear on the SCA channels A & B in the same time window. The time difference is calculated by subtracting the starting points of the two square waves that originated from each detector.

Although the data view of the time difference histogram has no value in the final equations to calculate the Unquantum Effect Ratio, it hints that the coincidences we are gathering happen very close in time. Gamma-ray is supposed to travel between two closely oriented detectors at light-speed in a nanosecond or even less.

In practice, coincident clicks may differ from a nanosecond up to several nanoseconds at the application level. Using a more extended time window, up to 300-1000ns, is good on the other hand because it will reveal random coincidences and also if there is something unexpected going on in the signal.

The resolution of the PicoScope model I have is 2ns when using all four channels (1ns when using two channels), so that is the best precision I can get to the coincident clicks. A picoscale accuracy (or several detectors) would be required to ensure the direction of the detections. Still, the nanoscale precision in this experiment is enough to tell if the clicks are coincident.

Many electrical devices are used in the experiment, starting from the scintillators and photomultipliers, SCA modules, BNC plus USB cables, that cause noise and time-shift coming to the system. The cables and amplifiers give a constant time-shift, which is easy to adjust in the graphs.

The jitter in time from the scintillator and PMT is expected if there was a true coincidence source (a source that emits two gammas in a single decay). We ensure that this is not the case with the isotope sources used in the experiment by the true coincidence test.

*Note: Tandem Piercer Experiment does not strictly speaking require the SCA module and square waves. The same could be achieved with a two-channel oscilloscope by triggering raw pulse signals and analyzing pulse peaks, for example, with the `scipy.find_peaks` algorithm. The center of the peaks could be compared, and their positions subtracted in order to get the time difference. But as said before, I rely on the standard SCA NIM module and its square wave, which seems to make things more robust.*

### Step 1.2 - Measure background rate (detectors far from each other)

Measurements in steps 1.2 - 1.4 are optional. They are meant to measure the general background radiation from the cosmic rays/beams and test the detectors. But, steps also have an important role on confirming that we are doing the right thing in the experiment. If detectors are far from each other, they should rarely produce any coincident clicks. A coincident click means that there is a cosmic ray that hits both detectors in an angle and direction that is same for both detectors. If there is a significant amount of coincident clicks when detectors are far from each other, it means that click-events have a correlation or that they would be in an entangled relation by quantum vocabulary.

While it is certainly possible that there can be two different cosmic rays that causes a click very close in time (in a few nanoseconds) in both detectors, they happen rarely due to a small volume of the detectors. The volume of the detector affects to the measurement, so I should note that both detectors' scintillation crystals, that I use, are two cubic centimeters in volume. When detectors are set one feet away from each other, I get coincident clicks roughly once in an hour. The rate of the coincidences should change according to the inverse-square law:

\begin{equation}
\text{rate} \ \propto \ \frac{1}{\text{distance}^2} \
\end{equation}

Measurements are made with a full spectrum because limiting spectrum to the tiny gamma region as described previosly, the coincident rate becomes even more rare. This is the reason I am not running this measurement but occasionally.

## Step 2 - Measure chance rate

After the first step I am ready to proceed to steps 2, 3, and 4. I will use the same gamma region settings that I set up in the previous step. The time window and timebase settings will differ on each measurement.

In chance rate measurement I have placed the isotope source in the front of the first detector. The counts of the gamma detections are collected in a maximum size buffer so that I get as a good count rate of clicks as possible.

The final rate is an average of the gamma pulses in a buffer. The measurement has been run for {chance_rate_run_time} seconds to retrieve the average values. Values and buffer size in seconds are stored to the `{change_rate_data.csv}` file. Chance rate $R_{c}$ is calculated by detector 1 rate multiplied by detector 2 rate multiplied by time window to be used in the background and tandem coincident measurement steps 3 and 4. Full equation for chance rate is:

\begin{equation}
R_c = R_1 R_2 T_w \\
R_c = 1 1 1 \\
R_c = 111
\end{equation}
