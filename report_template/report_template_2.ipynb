# Tandem Piercer Experiment - Report

## Intro

This fundamental physics experiment was originally invented by American electrical engineer Eric Reiter.

Tandem Piercer Experiment (TPE) is based on Reiter's long research and study of semiclassical electron model, so-called threshold model (also called accumulation hypothesis or loading theory). TPE demonstrates how two scintillation detectors placed in tandem geometry together with gamma-ray sources can produce an effect which may either support the threshold model or indicate an unexplained anomaly in fundamental physics.

TPE resembles the beam split coincidence experiment in certain ways. TPE also has implications in the wave-particle duality theory. The loading theory for electrons was first proposed by Max Planck (now known as his second theory) and contemporary physicists. Later on, the loading theory was abandonded by the very same scientists, which caused that the loading theory was rejected and eventually forgotten in the mainstream physics. TPE experiment and its result raises those previous ideas in the daylight again and forces us to ask, if there is still room for the semiclassical interpretation in fundamental physics.

### Contents

This report contains data tables and graphs, tables, and explanations from each measurement steps in TPE. The first step (1) is to detect the gamma-ray photopeak from the spectrum. The second step (2) is to measure the chance rate of the gamma-ray photopeaks. The third step (3) is to measure the background rate. The fourth and the final step (4) is to measure coincidences of the gamma-ray photopeaks in the tandem geometry of the detectors.

The final result is called Unquantum Effect Ratio or UQE in short. Unquantum-term is used because the chance rate measured in the seconds step shows the statistical prediction of the coincident gamma peaks according to quantum mechanics. The result of the Tandem Piercer Experiment, on the other hand, yields a correlation between the detectors, not just random distribution of coincidences equal to chance. If UQE is bigger than one, it means that the coincidence rate of gamma-ray photo peaks exceeds the chance rate. This should not happen in the quantum world, because the electron ionization in the scintillator produced by a single gamma-ray is expected to consume all of its energy in the process. But, for some reason, we often witness that is not the case in the Tandem Piercer Experiment. There seems to be more coincidences between the detectors than expected. How could that be explained? I hope that by these reports we can gather more data and understanding what is happening in the experiment.

## Step 1 - Detect Gamma-ray photopeak

This is a crusial part of the experiment. We need to produce the pulse height spectrum of the selected isotope and pinpoint the gamma-ray photopeak (later called just gammas) from it. We rather think gammas as clicks in the detector without commiting too much, at this point, what is the physical ontology of the gamma-ray in the electromagnetic spectrum.

By the previous work of scientists in the history, we know that gammas reside in the certain energy level (keV) for specific isotopes. Reiter has found two isotopes which can be used in the Tandem Piercer Experiment, namely `Cadmium 109` and `Cobalt 57`. They have the most clear and distinguished photopeaks in the NaI(Tl) scintillators. Corresponding gammas for `Cadmium 109` and `Cobalt 57` can be found from `88keV` and `122keV` respectively (https://www.cpp.edu/~pbsiegel/bio431/genergies.html). Other criteria among the clear photopeak is the half-life time, which also must be some reasonable value to make the experiment possible. Half-lifes for `Cadmium 109` and `Cobalt 57` are `453 days` and `279 days`, which are suitable for doing the experiment over a few years.

In this experiment the following source is used:

<h3 style="color:red; padding-bottom:15px">{pulse_source}</h3>

The first part is the name of the isotope. The second part is the activity value in microcuries (μCi). 1μCi means 3.70 × 10<sup>4</sup> disintegrations per second which is another way of denoting the radionuclide activity in Becquerels (Bq). Examption values declared by STUK limits that, as a consumer, I can use radionuclides up to 10μCi (10<sup>6</sup>Bq to be precise) to handle it safely. Radiation with 10μCi source is practically zero when measuring radiation ten centimeters from the source. Thus, the isotope source needs to be set very close to the detectors, within few centimeters.

The reference shape of the spectrum for {pulse_source} is shown below.

Pulse height spectrum is made of getting maximum value of the pulses retrieved from the oscilloscope which corresponds to channel number in the above picture. Maximum values are then collected and plotted to the graph which shows how certain values start to cumulate as peaks in the graph. In my graphs, I use Volts in x-axis which are converted from ADC digital values provided by PicoScope. So, in practice there are three common ways of denoting the x-axis: channel numbers, Volts, and ADC. Also the keV could be used if the gamma photopeak is defined to its location in the X-axis.

*Note: You should not confuse pulse height spectrum channel-term with the A, B, C, D channels I'm using to refer the four different signal channels in the oscilloscope.*

There are also Multichannel Analyzers (MCA) devices that could do the pulse height spectrum. I have done the same with Python by getting max value of the each signal data buffer when PicoScope has triggered and found a pulse coming from the raw signal of the detector. In practice, getting the pulse height spectrum in my set up is a combination of triggering the SCA square wave pulse and at the same time retrieving data from the corresponding raw pulse signal.

SCA module detects the pulse from the detector and creates a high precision square pulse. The square pulse height and width provided by SCA module are always same, while the curve of the pulses in raw signal has various shape, length and height. In my experiment set up, I rely on the SCA nuclear instrumentation (NIM) module, because they are dedicated on analyzing gamma- and x-ray pulses. SCA provides a square pulse with high precision up to picoscale which means that I have a good starting point on measuring the distance between coincident pulses between the two different channels (red Channel A & green Channel B in the picture).

### Regarding the resolution and time window

In the following measurements I'm going to need this information to generate a time difference histogram to see, that coincidences have centered peak in the graph. The time difference is measured by getting the starting point of the square wave when they appear on both channels A & B, and then taking a time difference between the starting points. Althought the view of the time difference histogram has no value in the final equations to calculate the Unquantum Effect Ratio, it gives a hint that the coincidences we are gathering, happens very close in time. Gamma-ray is supposed to travel between two closely oriented detectors in light-speed in a nanosecond or even less. The resolution of the PicoScope model I have, is 2ns when using all four channels (1ns when using two channels), so that is the best precision I can get to the coincident clicks. In the reality, since there are many electrical devices starting from the scintillators and photomultipliers, BNC plus USB-cables, and noise coming to the system from various places, we deduce that coincident clicks may differ from less than a nanosecond up to 100 nanoseconds in the application level. Using a longer time window, up to 300-1000ns is good on the other hand, because it will reveal random coincidences and also, if there is something unexpected going on in the signal.

NOTE: Tadem Piercer Experiment does not strictly speaking require SCA module. Same could be achieved with two channel oscilloscope by analyzing peaks in the raw data for example with `scipy.find_peaks` algorithm. But as said, I rely on the SCA NIM module, which seems to make things more robust.

### SCA and Region settings

Let us retrieve the region and SCA module settings from the saved configuration file and print them for a convenience:

```python
import json, pandas as pd

region = 'gamma' # gamma|full

with open('regions.json') as json_file:
    data = json.load(json_file)

def format_channel_data(data, channel_name):
    limit_index = 0 if channel_name == 'channel_a' else 1
    channel = {
        'Spectrum_ADC_low_limit': data['spectrum_low_limits'][limit_index], 
        'Spectrum_ADC_high_limit': data['spectrum_high_limits'][limit_index] 
    }
    channel.update(data['sca_module_settings'][channel_name])
    return {key.title().replace('_', ' '): value for key, value in channel.items()}

pd.DataFrame({
    'Channel A': format_channel_data(data[region], 'channel_a'),
    'Channel B': format_channel_data(data[region], 'channel_b')
})
```

## Step 2 - Measure chance rate

After the first step I am ready to proceed to steps 2, 3, and 4. I will use same gamma region settings that I set up in the previous step. The time window and timebase settings will differ on each measurement.

In chance rate measurement I have placed the isotope source in the front of the first detector. The counts of the gamma detections are collected in a maximum size buffer so that I get as good count rate of clicks as possible.

Th final rate is an average of the gamma pulses in a buffer. The measurement has been run for {chance_rate_run_time} seconds to retrieve the average values. Values and buffer size in seconds are stored to the `{change_rate_data.csv}` file. Chance rate R<sub>c</sub> is calculated by detector 1 rate multiplied by detector 2 rate multiplied by time window to be used in the background and tandem coincident measurement steps 3 and 4. Full equation for chance rate is:

R<sub>c</sub> = R<sub>1</sub> * R<sub>2</sub> * T<sub>w</sub>
